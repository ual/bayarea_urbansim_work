```python
import numpy as np
import pandas
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
a = orca.get_injectable('unit_aggregations')[0].to_frame()[['residential_price','sqft_per_unit','zone_id']]
```

```python
a.head()
```

```python
a['zone_id'].describe()
```

```python
b = orca.get_injectable('unit_aggregations')[2].to_frame()[['autoPeakTotal', 'transitPeakTotal' , 'autoOffPeakRetail']]
```

```python
b.head()
```

```python
b['id'] = b.index
```

```python
import pandas as pd
df=pd.merge(a, b, left_on='zone_id', right_on='id')
```

```python
df.head()
```

```python
model = ols(" np.log1p(residential_price)~ np.log1p(sqft_per_unit)+ autoPeakTotal + transitPeakTotal + autoOffPeakRetail", data=df).fit()
```

```python
print model.summary()
```

```python
c = model.resid
print type(c)
print len(c)
print len(a)
```

```python
bb = orca.get_injectable('unit_aggregations')[0]
orca.add_table('my_table', bb.to_frame())
orca.add_column('my_table', 'residual', c)
```
